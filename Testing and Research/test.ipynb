```python
import pandas as pd
from Bio import SeqIO
import os
```

```python
DATASET_PATH = "Covid India Dataset/"
os.listdir(DATASET_PATH)
```

```python
# open fasta files
fasta_sequences = []
for fasta in os.listdir(DATASET_PATH):
    if fasta.endswith(".fasta"):
        for seq_record in SeqIO.parse(DATASET_PATH + fasta, "fasta"):
            fasta_sequences.append(seq_record)
```

```python
for seq_record in fasta_sequences:
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
    print("")
```

```python
# print metadata
metadata = pd.read_csv(DATASET_PATH + "metadata.csv")
metadata.head()
```

```python
# metadat should only contain accession id, related id and location
metadata = metadata[["Accession ID", "Related ID", "Location", "Lineage", "Sample Collection Date"]]
metadata.head()
```

```python
# if accession id starts with EPI_ISL_ then replace entire id with empty string
metadata["Accession ID"] = metadata["Accession ID"].apply(lambda x: "" if str(x).startswith("EPI_ISL_") else str(x))

# if related id starts with EPI_ISL_ then replace entire id with empty string
metadata["Related ID"] = metadata["Related ID"].apply(lambda x: "" if str(x).startswith("EPI_ISL_") else str(x))
```

```python
# if accession id is empty then replace it with related id
metadata["Accession ID"] = metadata.apply(lambda x: x["Related ID"] if x["Accession ID"] == "" else x["Accession ID"], axis=1)
```

```python
# remove related id column
metadata = metadata[["Accession ID", "Location", "Lineage", "Sample Collection Date"]]
metadata.head()
```

```python
# if accession id ends with .1 then remove .1
metadata["Accession ID"] = metadata["Accession ID"].apply(lambda x: x[:-2] if str(x).endswith(".1") else str(x))
```

```python
# rename accession id column to id
metadata = metadata.rename(columns={"Accession ID": "ID"})
```

```python
metadata.head()
```

```python
metadata["Location"] = metadata["Location"].apply(lambda x: x.split("/")[-1].strip())
metadata.head()
```

```python
fasta_seq = []
for seq_record in fasta_sequences:
    fasta_seq.append([str(seq_record.id), str(seq_record.seq)])

fasta_seq = pd.DataFrame(fasta_seq, columns=["ID", "Sequence"])
fasta_seq.head()
```

```python
# merge metadata and fasta_seq
merged = pd.merge(metadata, fasta_seq, on="ID")
merged.head()
```

```python
# save merged dataframe to csv
merged.to_csv(DATASET_PATH + "merged.csv", index=False)
```

## Lineage Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plot count of each lineage
lineage_count = merged["Lineage"].value_counts()
# keep only top 50 lineages
lineage_count = lineage_count[:50]
plt.figure(figsize=(20, 10))
sns.barplot(x=lineage_count.index, y=lineage_count.values)
plt.xticks(rotation=90)
plt.xlabel("Lineage")
plt.ylabel("Count")
plt.title("Count of each lineage")
plt.show()
```

```python
merged = pd.read_csv(DATASET_PATH + "merged.csv")
merged.head()

# group Surat, Vadodara, Rajkot, Gandhinagar, Ahmedabad as Gujarat
merged["Location"] = merged["Location"].apply(lambda x: "Gujarat" if x in ["Surat", "Vadodara", "Rajkot", "Gandhinagar", "Ahmedabad"] else x)

# group Western Uttar Pradesh, Uttar Pradesh as Uttar Pradesh
merged["Location"] = merged["Location"].apply(lambda x: "Uttar Pradesh" if x in ["Western Uttar Pradesh", "Uttar Pradesh"] else x)

# group Delhi, New Delhi as Delhi
merged["Location"] = merged["Location"].apply(lambda x: "Delhi" if x in ["New Delhi", "Delhi"] else x)
```

```python
lineage_count = lineage_count[lineage_count > 30]

# group lineages by location
lineage_by_location = merged.groupby("Location")["Lineage"].value_counts()
lineage_by_location = lineage_by_location.unstack()
lineage_by_location.head()

# only keep lineages that are in lineage_count
lineage_by_location = lineage_by_location[lineage_by_location.columns.intersection(lineage_count.index)]

# set nan values to 0
lineage_by_location = lineage_by_location.fillna(0)

# convert float values to int
lineage_by_location = lineage_by_location.astype(int)

locations = ["Delhi", "Gujarat", "Karnataka", "Maharashtra", "Uttar Pradesh", "West Bengal"]

# keep locations that are in locations list
lineage_by_location = lineage_by_location[lineage_by_location.index.isin(locations)]

# remove unclassified lineage
lineage_by_location = lineage_by_location.drop("unclassifiable", axis=1)
lineage_by_location = lineage_by_location.drop("Unassigned", axis=1)

# plot heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(lineage_by_location, annot=True, fmt="d")
plt.xticks(rotation=90)
plt.xlabel("Lineage")
plt.ylabel("Location")
plt.title("Lineage by location")
plt.show()
```

```python
# group dates by month and year
merged["Sample Collection Date"] = pd.to_datetime(merged["Sample Collection Date"])
merged["Month"] = merged["Sample Collection Date"].dt.month
merged["Year"] = merged["Sample Collection Date"].dt.year

merged.head()
```

```python
# group quarterly
merged["Quarter"] = merged["Month"].apply(lambda x: 1 if x in [1, 2, 3] else 2 if x in [4, 5, 6] else 3 if x in [7, 8, 9] else 4)
```

```python
merged.head()
```

```python
# group lineages by month and year
lineage_by_date = merged.groupby(["Year", "Quarter"])["Lineage"].value_counts()
lineage_by_date = lineage_by_date.unstack()
lineage_by_date.head()

# only keep lineages that are in lineage_count
lineage_by_date = lineage_by_date[lineage_by_date.columns.intersection(lineage_count.index)]

# set nan values to 0
lineage_by_date = lineage_by_date.fillna(0)

# convert float values to int
lineage_by_date = lineage_by_date.astype(int)

# remove unclassified lineage
lineage_by_date = lineage_by_date.drop("unclassifiable", axis=1)
lineage_by_date = lineage_by_date.drop("Unassigned", axis=1)

plt.figure(figsize=(20, 10))
sns.heatmap(lineage_by_date, annot=True, fmt="d")
plt.xlabel("Lineage")
plt.ylabel("Year, Quarter")
plt.title("Lineage by date")
plt.show()
```
